```python
pakistan_data = data[data["Country"] == 'Pakistan']
# Normalizing = "making range of each column same"
# Applying formula of normalizing value on each cell of each column 
for column in pakistan_data.columns:
    if column != 'Country':
        exact_column = pakistan_data[column]
        maxi = exact_column.max()
        mini = exact_column.min()
        for index, value in exact_column.iteritems():
            try:
                normalize_value = (value - mini)/(maxi - mini)
            except:
                normalize_value = (value - mini)
            pakistan_data.at[index, column] = normalize_value

for col in pakistan_data.columns:
    plt.plot(years, pakistan_data[col])

```
